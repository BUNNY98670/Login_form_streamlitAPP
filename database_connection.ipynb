```python
## step3 : connect database

conn = connector.connect(
    host="localhost",
    user="root",
    password="bunnyspike246801",
    database="Student_db"
    )

cursor = conn.cursor()
print("Connected to database :", conn.is_connected())
```

##step4 : create table


```python
cursor.execute("CREATE TABLE IF NOT EXISTS Student_info (id INT AUTO_INCREMENT PRIMARY KEY, name VARCHAR(255), age INT)")
print("Table 'Student_info' created successfully.")
```

```python
STEP5: INSERT VALUES
```

```python
cursor.execute("INSERT INTO Student_info (name, age) VALUES ('Alice', 22)")
print("Record inserted successfully.")
```

```python
step 6:MULTIPLE VALUES
```

```python
#insert more records as needed

sql = "INSERT INTO Student_info (id,name, age) VALUES (%s, %s, %s)"
val = [
    (2,'Bob', 23),
    (3,'Charlie', 24),
    (4,'David', 25)
    ]
cursor.executemany(sql, val)
conn.commit()
print(cursor.rowcount, "records inserted successfully.")
```

DISPLAY TABLE


```python
cursor.execute("SELECT * FROM Student_info")
records = cursor.fetchall()

for row in records:
    print(row)
```
